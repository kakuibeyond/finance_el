## 金融新闻文本可视化分析
- 新闻关键词分布
- 新闻文本长度分布

```python
import pandas as pd
df2=pd.read_csv('fin_articles.csv',encoding='utf8')
df2.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(dpi=100, figsize=(20, 8))
sns.set_style('whitegrid',{'font.sans-serif':['simhei','Arial']})
sns.countplot(y=df2['keywords'])
```

```python
sns.set(style='dark')
sns.set_style("dark", {"axes.facecolor": "#e9f3ea"})
plt.figure(dpi=100, figsize=(20, 8))
g=sns.distplot(df2['len_text'],
               hist=True,
               bins=500,#修改箱子个数
               kde=False,
               color="#098154")
g.set_xlim(0,20000) # 限制x的值为[0,20]
```
